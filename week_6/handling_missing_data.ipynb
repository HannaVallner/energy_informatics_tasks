````python
# Imports
import datetime
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
````

````python
weather = pd.read_csv('../data/weather_2022.csv',  delimiter=';', decimal='.', index_col = False)
weather['timestamp'] = pd.to_datetime(weather['Local time in Tallinn (airport)'], dayfirst=True)
weather.set_index('timestamp', inplace=True)
weather.drop(columns=['Local time in Tallinn (airport)'], inplace=True)
````

````python
missing_count = weather.isna().sum()
missing_percent = (missing_count / len(weather)) * 100

columns_to_keep = missing_percent[missing_percent < 20].index
weather_cleaned = weather[columns_to_keep].copy()

numeric_columns = weather_cleaned.select_dtypes(include="number").columns
weather_cleaned[numeric_columns] = weather_cleaned[numeric_columns].interpolate(
    method="linear"
)
weather_cleaned[numeric_columns] = weather_cleaned[numeric_columns].ffill()
weather_cleaned[numeric_columns] = weather_cleaned[numeric_columns].bfill()
````

````python
print("Missing values before cleaning:", weather.isna().sum().sum())
print("Missing values after cleaning:", weather_cleaned.isna().sum().sum())
removed_columns = missing_percent[missing_percent >= 20].index
print("Removed columns:", removed_columns)
````
